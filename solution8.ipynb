如下图为我校校园网上新闻网页中的一区域，对应的部分html网页代码在news.txt文本文件中，请用记事本打开此文件，将所有内容作为一个字符串数据，在Python编程环境中赋于一个字符串变量，再使用正则表达式提取出14条新闻的标题、及后方的发布的日期

```python
import re
txt=open(r'.\news.txt',encoding='utf-8').read()
pat1=r'<div id="wp_news_w9">(.+?)</ul>'
txt1=re.search(pat1,txt,re.S).group(1)
pat2=r'<li.+?<a.+?>(.+?)</a>.+?<span.+?>(.+?)</span>'
lst=re.findall(pat2,txt1,re.S)
lst[:5]
```

```python
i=1
for c,d in lst:
    print(f"新闻标题{i}：{c}\n发布时间{i}：{d}")
    print()
    i=i+1
```

如右图所示为一成都美食网页，对应的html网页数据存放在”成都美食.txt”文件中，用记事本打开此文件，将所有内容作为一个字符串数据，在Python编程环境中赋于一个字符串变量，再使用正则表达式提取出六个“美食名称”、去哪里吃的“店铺名称”。

```python
import re
txt=open(r'.\扬州美食.txt',encoding='utf-8').read()
pat1=r'<li.+?<dt>(.+?)</dt>\s+<dd>(.+?)<.+?"ellipsis">\s+<a.+?>(.+?)</a>'
tlst=re.findall(pat1,txt,re.S)
for name,content,addr in tlst:
    print(f"美食名称：{name}\n美食介绍：{content}\n店铺名称：{addr}")
    print()
```

```python
pat2=r'<li.+?<dt>(.+?)</dt>\s+<dd>(.+?)<.+?<dd.+?<p.+?<a.+?>(.+?)</a>、<a.+?>(.+?)</a>、<a.+?>(.+?)</a>、<a.+?>(.+?)</a>'
yzlst=re.findall(pat2,txt,re.S)
for da1,da2,da3,da4,da5,da6 in yzlst:
    print(f"美食名称：{da1}\n美食介绍：{da2}\n店铺名称：{da3}、{da4}、{da5}、{da6}\n")
```

用记事本打开“五哈弹幕.txt”高维数据文件，将所有内容作为一个字符串数据，在Python编程环境中赋于一个字符串变量，再使用正则表达式提取出每个弹幕的评论(content后面的内容)及评论的ID号(commentid后面的数字)。

```python
txt=open(r'.\五哈弹幕.txt',encoding='utf-8').read()
pat=r'"commentid":"(\d+)","content":"(.+?)"'
tlst=re.findall(pat,txt,re.S)
for id,conc in tlst:
    print(f"评论ID：{id}, 评论内容：{conc}")
```

用记事本打开employer.txt文件，将整个内容通过复制、粘贴，赋值给一个字符串变量，再使用正则表达式，分别提取出所有员工的姓名、出生日期，电话号码，电子邮箱等信息。

```python
txt=open(r'.\employer.txt',encoding='utf-8').read()
txt
```

```python
pat=r'\[(.+?)]'
ls=re.findall(pat,txt,re.S)
lst=[eval(w) for w in ls]
for name,sex,birth,tele,email,addr in lst:
    print(f"姓名:{name}, 性别:{sex}, 出生日期:{birth}, 手机:{tele}, 邮箱:{email},住址:{addr.split()[0]}，邮编：{addr.split()[-1]}")
```

用记事本打开zlgwfzzp.json文件，将整个内容通过复制、粘贴，赋值给一个字符串变量words，再使用正则表达式，分别提取所有招聘单位的单位名称、招聘岗位名称、薪水、工作城市名及岗位待遇信息，并按下列格式每五行输出一个单位的招聘信息然后空一行，再输出下一个单位的信息。

```python
import re
words=open(r'.\zlgwfzzp.json',encoding='gbk').read()
words
```

```python
pat=r'{"number".+?"company".+?"name": "(.+?)".+?"salary": "(.+?)".+?"jobName": "(.+?)".+?"city".+?"name": "(.+?)".+?"welfare": \[(.*?)]'
lst=re.findall(pat,words,re.S)
for i in range(len(lst)):
    print("单位名称：{}\n薪水：{}\n岗位名称：{}\n工作城市：{}\n岗位福利：{}".format(lst[i][0],lst[i][1],lst[i][2],lst[i][3],lst[i][4]))
    print()
```
