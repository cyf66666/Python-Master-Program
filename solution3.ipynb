通过列表推导式，使用random库中的choice()函数，从[1,20]中随机生成50个自然数的列表，输出列表中最大值所对应的索引下标。(提示：最大值可能会重复多个，参考上课案例，可使用count()和index()两个函数循环求解，也可使用enumerate()函数遍历序列解包求解)

```python
import random
ls=[random.choice(range(1,21)) for i in range(50)]
m=max(ls)
print(ls,"最大值：",m)
##用列表的下标索引
for i in range(len(ls)):
    if ls[i]==m:
        print(i,end='\t')
print()
##上述方法的列表推导式        
print([i for i in range(len(ls)) if ls[i]==m])

##用enumerate()函数
for i,v in enumerate(ls):
    if v==m:
        print(i,end='\t')
print()
##上述方法的列表推导式
print([i for i,v in enumerate(ls) if v==m])

##使用count()和index()函数
n=ls.count(m)
ind=0
for i in range(n):
    ind=ls.index(m,ind)
    print(ind,end='\t')
    ind+=1
    
```

通过列表推导式，使用random库中的choice方法，使用string库中ascii_letters英文字符串, 从52个大小写英文字符串中随机抽取10000个字符，用字典统计出每个字符被抽取的次数，并输出出现次数最多的前五个字母及出现的次数。(提示：可使用字典的get()方法求解字符出现的次数，也可以使用字典推导式生成，然后再对字典进行排序求解)

```python
import string,random
ls=[random.choice(string.ascii_letters) for i in range(10000)]
##用字典计数统计
dic1={}
for ch in ls:
    dic1[ch]=dic1.get(ch,0)+1
print(dic1)  
print()
##用字典推导式
dic2={k:ls.count(k) for k in set(ls)}
print(dic2)
print()
##用第三方库Counter
from collections import Counter
word="".join(ls)
dic3=Counter(word)
print(dic3)
```

```python
for c,n in sorted(dic3.items(),key=lambda x:x[1],reverse=True)[:5]:
    print("字母：{0}\t次数：{1}".format(c,n))
```

某学院有四个年级(2020,2021,2022,2023)，每个年级分别有“临床医学”、“儿科学”和“影像学”三个专业，每个专业有三个自然班(01,02,03)，每个自然班有30名学生，共计1080名学生，每个学生的信息包括：学号(stid由10位数字构成，1-4位是入学年度，5-6位是学院专业编号，7-8位是同年级专业班级编号，9-10位是班级学生的序号，如“2021110125”)、姓名(stname，如“张国庆”)、班级名(bjname，如“临床医学2021-02班”)，电话(phonenum, 如“13801239811”)及高等数学(math)、英语(eng)、计算机(compu)三门课程的成绩。
数据源：要求导入faker及random库，通过列表推导式生成学号、姓名、班级名，电话，三门课程的成绩可使用random库是的子库从[45,100]范围内随机抽取。
(1)	使用zip()函数将学号(stid)、姓名(stname)、班级名(bjname)、电话(phonenum)及高等数学(math)、英语(eng)、计算机(compu)三门课程的成绩，共七个数据集成到一个zip对象容器中，为了便于操作，可转为列表对象stinfomation。
(2)	列表中包含着1080名的学生信息，每个学生信息的7个数据保存在一个元组对象中。若要再为每个学生信息增添一个总分(total)数据，其值是三门课程成绩之和，请编程实施。(提示：可参考上课讲解，先求解每个学生的三门课程总分，再将每个学生的总分加入到学生信息的元组中)
(3)	若要按总分升序对列表中的学生信息进行排序，根据排序结果，输出总分最高的前三名学生的学号、姓名、班级名和总分四项数据，每名学生的数据占一行。
输出格式：学号：xxxxxxxxxx  姓名：xxxxx  班级名：xxxxxxxx  总分：xxx
(4)	如对总分进行降序排序时，总分相同，再按math成绩降序排列，输出总分最低的前三名学生的姓名、高等数学和总分数据，输出格式同上。
(5)	思考：能否用切片方法求解出每个班级总分最高(或最低)的学生信息数据？

```python
import faker
from random import *
fak=faker.Faker("zh-cn")         #创建一个fak对象
#使用列表推导式生成四个年级，每个年级三个专业，每个专业三个班级，每个班级30人的学号
stid=[str(x*10**6+y*10**4+z*10**2+n) for x in range(2020,2024) for y in range(11,14) for z in range(1,4) for n in range(1,31) ]
#使用列表推导式生成总人数1080人的姓名
stname=[fak.name() for i in range(1080)]
#使用列表推导式生成总人数1080人的电话号码
phonenum=[fak.phone_number() for i in range(1080)]
zy=['临床医学','儿科学','影像学']        #创建一个包含三个专业名称的列表
#使用列表推导式生成四个年级，每个年级三个专业，每个专业三个班级，每个班级30人的班级名称
bjname=["{1}{0}-{2}班".format(x,y,z) for x in range(2020,2024) for y in zy for z in range(1,4) for i in range(30)]

#分别生成1080人三门课程的随机成绩
math=[randint(45,100) for i in range(1080)]
eng=[randrange(45,101) for i in range(1080)]
compu=[choice(range(45,101)) for i in range(1080)]
```

(1)	使用zip()函数将学号(stid)、姓名(stname)、班级名(bjname)、电话(phonenum)及高等数学(math)、英语(eng)、计算机(compu)三门课程的成绩，共七个数据集成到一个zip对象容器中，为了便于操作，可转为列表对象stinfomation。

```python
#使用zip函数将以上数据组合成1080名学生的信息，转换成列表数据，便于操作
stinfo=list(zip(stid,stname,bjname,phonenum,math,eng,compu))
#使用列表切片，查看前10名学生信息
stinfo[:10]
```

(2)	列表中包含着1080名的学生信息，每个学生信息的7个数据保存在一个元组对象中。若要再为每个学生信息增添一个总分(total)数据，其值是三门课程成绩之和，请编程实施。(提示：可参考上课讲解，先求解每个学生的三门课程总分，再将每个学生的总分加入到学生信息的元组中)

```python
#对列表中每个学生的元组数据，再"追加"一个三门课程的成绩和的总分信息total，保存在新的列表中
stin=[]                      #设置一个存放每个学生包含三门课程成绩之和数据的新列表
for w in stinfo:             #从列表中遍历每个学生信息的元组
    st=w+(sum(w[-3:]),)       #使用元组切片，获得学生的三门课程成绩，再求和，然后放在元组中，和每个学生信息的元组相"加"，生成一个新元组
    stin.append(st)          #将生成的新元组追加到新列表中
stin[:10]                    #查看新列表的前10名学生信息
```

```python
#以上算法可使用列表推导式
stin1=[w+(sum(w[-3:]),) for w in stinfo]
stin1[:10]
```

```python
#也可将原列表中的每个学生信息的元组数据先转换中列表数据，再使用列表的append()方法，将每个学生的三门课程成绩之和追加上去
stinls=[list(da) for da in stinfo] 
stin2=[]
for da in stinls:
    da.append(sum(da[-3:]))
    stin2.append(da)
stin2[:10]
```

(3)	若要按总分升序对列表中的学生信息进行排序，根据排序结果，输出总分最高的前三名学生的学号、姓名、班级名和总分四项数据，每名学生的数据占一行。
输出格式：学号：xxxxxxxxxx  姓名：xxxxx  班级名：xxxxxxxx  总分：xxx

```python
#按总分total降序排列，最高分排在最后一位，使用sorted()中的key参数，设置排序的规则，lambda匿名函数中，
#冒号前的x代表可迭代列表对象中的每一个元组对象，冒号x[7]表示以这个元组对象中的下标索引为7的数据作为操作对象，进行排序
stsort=sorted(stin,key=lambda x:x[7],reverse=False)    
for da in stsort[-3:][::-1]:        #对列表的后三位再倒序，使得最高分在第一位，次高分在第二位...
    print("学号：",da[0],"\t姓名：",da[1],"\t班级名称：",da[2],"\t总分：",da[7])     #da数据对象是一个元组，可以用下标索引访问每个数据
```

(4)	如对总分进行降序排序时，总分相同，再按math成绩降序排列，输出总分最低的前三名学生的姓名、高等数学和总分数据，输出格式同上。

```python
#sorted()函数的key参数可设置排序的规则，lambda匿名函数中冒号前的x是指可迭代窗口中的每一个数据对象，
#冒号后面则是指对每一个数据对象按什么规则来操作，可以把多个操作要求放在圆括号里作为一个元组对象，元组是有序的，有前后次序。
st_sort=sorted(stin,key=lambda x:(x[7],x[4]),reverse=True)
st_sort[:10]
```

(5)	思考：能否用切片方法求解出每个班级总分最高(或最低)的学生信息数据？

```python
#可以对学生信息列表，先按班级名称排序，同一班级的学生再按总分排序，然后通过改变步长的切片，获取每个班级最高分或最低分的学生信息
stlst=sorted(stin,key=lambda x:(x[2],x[7]))
st_min=stlst[::30]
for st in st_min:
    print("班级名称：{0}\t总分最低分：{1}".format(st[2],st[7]))
print()    
st_max=stlst[29::30]
for st in st_max:
    print("班级名称：{0}\t总分最高分：{1}".format(st[2],st[7]))
```
